```python
import numpy as np
import pylab as plt

def time(T):
    return arange(0.0001,0.4,1./200.)

def case(t,n):
    if n == 0:
        func = 10.*exp(2.*pi*1j*10.*t)
    if n == 1:
        func = 10.*exp(-2.*pi*1j*10.*t)
    if n == 2:
        func = 10.*exp(2.*pi*1j*10.*t + 1j*pi/2.)
    if n == 3:
        func = 10.*exp(2.*pi*1j*10.*t + 1j*pi/2.) + 5.*exp(2.*pi*1j*20.*t - 1j*pi/2.)
    if n == 4:
        func = 6.*cos(2.*pi*3.*t)
    if n == 5:
        func = 4.*cos(2*pi*3.*t + pi/4.)
    if n == 6:
        func = 3.*sin(2*pi*4.*t)
    if n == 7:
        func = 13.+ 7.*sin(2*pi*7.*t) + 3.*cos(2*pi*3.*t+ pi/4.)+ sin(2*pi*7.*t) + 7. + 3.*sin(2*pi*3.*t+ pi/4.)
    return func

tim = time(0)
```

```python
#Fourier transformation
def trans(f):
    M = len(f)
    transform = np.zeros_like(f,dtype="complex")
    for i in range(M):
        for j in range(M):
            transform[i] += f[j]*exp(-2.*pi*1j*j*i/(M))
    return transform
```
