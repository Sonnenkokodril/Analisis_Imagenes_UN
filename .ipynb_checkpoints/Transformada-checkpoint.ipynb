```python
import numpy as np
import pylab as plt

def time(T):
    return arange(0.0001,1.,1./200.)

def case(t,n):
    if n == 0:
        func = 10.*exp(2.*pi*1j*10.*t)
    if n == 1:
        func = 10.*exp(-2.*pi*1j*10.*t)
    if n == 2:
        func = 10.*exp(2.*pi*1j*10.*t + pi/2.)
    if n == 3:
        func = 10.*exp(2.*pi*1j*10.*t + pi/2.) + 5.*exp(2.*pi*1j*20.*t - pi/2.)
    if n == 4:
        func = 6.*cos(2.*pi*3.*t)
    if n == 5:
        func = 4.*cos(2*pi*3.*t + pi/4.)
    if n == 6:
        func = 3.*sin(2*pi*4.*t)
    if n == 7:
        func = 13.+ 7.*sin(2*pi*7.*t) + 3.*cos(2*pi*3.*t+ pi/4.)+ sin(2*pi*7.*t) + 7. + 3.*sin(2*pi*3.*t)
    return func

tim = time(0)
```

```python
#Fourier transformation
def trans(f):
    M = len(f)
    transform = np.zeros_like(f,dtype="complex")
    for i in range(M):
        for j in range(M):
            transform[i] += f[j]*exp(-2.*pi*1j*j*i/(M))
    return transform
```

```python
#Plot functions
plt.figure( figsize = (12,8))

plt.subplot(421)
plot(tim,case(tim,0).real, 'r-o', label='F1 Real')
plot(tim,case(tim,0).imag, 'b.-', label='F1 Imag')

plt.subplot(422)
plot(tim,case(tim,1).real, 'g-o', label='F2 Real')
plot(tim,case(tim,1).imag, 'b.-', label='F2 Imag')

plt.subplot(423)
plot(tim,case(tim,2).real, '-oc', label='F3 Real')
plot(tim,case(tim,2).imag, 'b.-', label='F3 Imag')

plt.subplot(424)
plot(tim,case(tim,3).real, '-om', label='F4 Real')
plot(tim,case(tim,3).imag, 'b.-', label='F4 Imag')

plt.subplot(425)
plot(tim,case(tim,4).real, '-oy', label='F5 Real')
plot(tim,case(tim,4).imag, 'b.-', label='F5 Imag')

plt.subplot(426)
plot(tim,case(tim,5).real, '-ok', label='F6 Real')
plot(tim,case(tim,5).imag, 'b.-', label='F6 Imag')

plt.subplot(427)
plot(tim,case(tim,6).real, '-oc', label='F7 Real')
plot(tim,case(tim,6).imag, 'b.-', label='F7 Imag')

plt.subplot(428)
plot(tim,case(tim,7).real, '-og', label='F8 Real')
plot(tim,case(tim,7).imag, 'b.-', label='F8 Imag')


plt.show()
```

```python
#Plot Fourier transformation
plt.figure(figsize = (12,8))
tim1=1./tim

plt.subplot(421)
plt.loglog(tim1,abs(trans(case(tim,0))), 'r-o', label='F1 Fourier Trans')

plt.subplot(422)
plt.loglog(tim1,abs(trans(case(tim,1))), 'g-o', label='F1 Fourier Trans')

plt.subplot(423)
plt.loglog(tim1,abs(trans(case(tim,2))), 'c-o', label='F1 Fourier Trans')

plt.subplot(424)
plt.loglog(tim1,abs(trans(case(tim,3))), 'm-o', label='F1 Fourier Trans')

plt.subplot(425)
plt.loglog(tim1,abs(trans(case(tim,4))), 'y-o', label='F1 Fourier Trans')

plt.subplot(426)
plt.loglog(tim1,abs(trans(case(tim,5))), 'k-o', label='F1 Fourier Trans')

plt.subplot(427)
plt.loglog(tim1,abs(trans(case(tim,6))), 'c-o', label='F1 Fourier Trans')

plt.subplot(428)
plt.loglog(tim1,abs(trans(case(tim,7))), 'g-o')

plt.show()
```
